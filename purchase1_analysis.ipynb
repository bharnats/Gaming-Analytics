```python
import pandas as pd
import os
import json
```

```python
file_path = os.path.join("purchase_data.json")
with open(file_path) as fileobj :
    data = json.load(fileobj)
purchase_df = pd.DataFrame(data)
purchase_df
```

```python
# Player count
player_count =len(purchase_df["SN"].value_counts())
print(player_count)
```

```python
# Number Of Unique Items
total_items = len(purchase_df["Item ID"].value_counts())
print(total_items)
```

```python





# Average Purchase Price
avg_price = round(purchase_df["Price"].mean(),2)
print(avg_price)
```

```python

# Total number of purchases (number of rows in the data frame)
len(purchase_df.index)
```

```python
# Total Revenue
total_revenue = round(purchase_df["Price"].sum(),2)
print(total_revenue)
```

```python
# gender Demographics
# Count of male players
male_df = purchase_df.loc[purchase_df["Gender"]=="Male",:]
male_count = len(male_df["SN"].unique())
print(male_count)
```

```python
# percentage of male players
total_count = len(purchase_df["SN"].unique())
percent_male =round((male_count/total_count)*100,2)
print(percent_male)
```

```python
# female players count
female_df = purchase_df.loc[purchase_df["Gender"]=="Female",:]
female_count = len(female_df["SN"].unique())
print(female_count)
# percentage of female players
percent_female = round((female_count/total_count)*100, 2)
print(percent_female)
```

```python
# Percentage of other/undisclosed gender players
other_count = total_count-male_count-female_count
print(other_count)
percent_other = round((other_count/total_count)*100,2)
print(percent_other)
```

```python
# purchase Analysis by gender
# purchase count -male
male_purchase = len(male_df)
print(male_purchase)
# purchase count -female
female_purchase = len(female_df)
print(female_purchase)
```

```python
# Average Purchase Price
# Male
avg_price_male =round((male_df["Price"].sum())/len(male_df["Price"]),2)
print(avg_price_male)
# Female
avg_price_female =round((female_df["Price"].sum())/len(female_df["Price"]),2)
print(avg_price_female)

```

```python
# Total Purchase Value
# Total purchase Value- Male
total_value_male = round(male_df["Price"].sum(),2)
print(total_value_male)
# Total Purchase Value -Female
total_value_female = round(female_df["Price"].sum(),2)
print(total_value_female)

```

```python
# Normalised Totals
# for male
norm_total_val_male = round((total_value_male/male_count), 2)
print(norm_total_val_male)
# for female
print(norm_total_val_male)
norm_total_val_female = round((total_value_female/female_count), 2)
```

```python
# age demographics
# find max and min of age column to fix the outer edges of the bins
max_age = purchase_df["Age"].max()
min_age = purchase_df["Age"].min()
print(max_age)
print(min_age)
```

```python
# creating bins
age_bins = [7,11,15,19,23,27,31,35,39,43,47]
age_labels = ["7 to 11","11 to 15","15 to 19","19 to 23","23 to 27","27 to 31","31 to 35","35 to 39","39 to 43","43 to 47"]
purchase_df["Age Summary"] = pd.cut(purchase_df["Age"], age_bins, labels=age_labels)
purchase_df.head()
```

```python
# purchase count by age bins
age_groupby = purchase_df.groupby("Age Summary")
purchase_count_bins = age_groupby["Age"].count()
purchase_count_bins
```

```python
# average Purchase Price within each bin
avg_price_bin = round(age_groupby["Price"].mean(),2)
avg_price_bin
```

```python
# total Purchase value
tot_pur_value_bin = round(age_groupby["Price"].sum(), 2)
tot_pur_value_bin
```

```python
# top 5 spenders
SN_groupby = purchase_df.groupby(purchase_df["SN"])
SN_groupby_df = pd.DataFrame(SN_groupby["Price"].sum())
SN_groupby_df = SN_groupby_df.reset_index()
SN_groupby_df.columns = ["SN","Amount_spent"]
#SN_groupby_df.sort_values("Amount_spent",ascending=False).head()
SN_groupby_df.head()
```

```python
# purchase count for top 5 spenders
pur_count_df = pd.DataFrame(purchase_df["SN"].value_counts())
pur_count_df = pur_count_df.reset_index()
pur_count_df.columns=["SN","Purchase Count"]

pur_count_df.head()

```

```python
# merging the amount spent and purchase_count tables  on SN
merge1_df = pd.merge(SN_groupby_df,pur_count_df,on = "SN")
merge1_df.head()
```

```python
SN_groupby_avg_df = pd.DataFrame(round(SN_groupby["Price"].mean(), 2))
SN_groupby_avg_df = SN_groupby_avg_df.reset_index()
SN_groupby_avg_df.columns = ["SN","Avg Purchase price"]
SN_groupby_avg_df.head()
```

```python
# merging the amount spent and purchase_count tables and average purchase Price on SN
top5_spenders_df = pd.merge(merge1_df,SN_groupby_avg_df,on = "SN")
top5_spenders_df = top5_spenders_df.sort_values("Amount_spent", ascending=False)

top5_spenders_df = top5_spenders_df.iloc[0:5,:]
top5_spenders_df
```
